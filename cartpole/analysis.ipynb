```python
def get_result(model_name):
    res_dir = pathlib.Path.cwd().joinpath('results')
    model_dir = res_dir.joinpath(model_name)
    model_stats = np.load(str(model_dir.joinpath('my_file.npy')), allow_pickle = True).item()
    return model_stats
```

```python
res1 = get_result('image_v_crop_128')
plt.plot(res1['time'])
plt.title('Time Vs Episode', fontsize = 22)
plt.xlabel('Episodes',fontsize = 15)
plt.ylabel('Time(sec)',fontsize = 15)

plt.show()

plt.plot(res1['reward'])
plt.title('Reward Vs Episode', fontsize = 22)
plt.xlabel('Episodes',fontsize = 15)
plt.ylabel('Reward',fontsize = 15)
plt.show()

print(f"""

    Min Reward: {min(res1['reward'])}
    Max Reward: {max(res1['reward'])}
    avg Reward: {sum(res1['reward'])/len(res1['reward'])}
    \n
    Avg time/Step: {round(np.average(res1['time']/res1['steps']),4)}
    Total Steps: {sum(res1['training_actions'].values())}
    Total training time: {round(sum(res1['time'])/60,4)}

""")
```
